```python
x = np.linspace(0, 100, 11)
y = np.linspace(0, 200, 11)

# Create a new figure object
fig = plt.figure()

# Add our axes to the figure
ax = fig.add_axes([0, 0, 1, 1])
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_title('Sample Graph')
ax.plot(x, y)
```
