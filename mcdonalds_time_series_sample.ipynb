```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

print ("Complete")
```

```python
# Read and view the mcdonalds csv data
mcdonalds_df = pd.read_csv('mcdonalds.csv', index_col='Date', parse_dates=True)
mcdonalds_df.head()
```

```python
# Plot data based on Adj. Close
mcdonalds_df['Adj. Close'].plot()
```

```python
# Let's zoom in on this chart and plot a specific date range: Jan 2007 - Jan 2009
mcdonalds_df['Adj. Close'].plot(xlim=['2007-01-01', '2009-01-01'], ylim=(30,50))
```

```python
# Let's grab a subset of our data frame
idx = mcdonalds_df.loc['2007-01-01': '2007-05-01'].index
stock = mcdonalds_df.loc['2007-01-01': '2007-05-01']['Adj. Close']
```

```python
# Plot this smaller data set
fig, ax = plt.subplots()
ax.plot_date(idx, stock, '-')
fig.autofmt_xdate()
```
